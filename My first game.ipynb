```python
import pygame
import time

WIN_WIDTH = 1024
WIN_HEIGHT = 600
BTN_WIDTH = 80
BTN_HEIGHT = 80
HP_WIDTH = 40
HP_HEIGHT = 40
FPS = 30
enemy_x = 30     #enemy的x座標
enemy_y = 230    #enemy的y座標

# color (RGB)    定義顏色RGB值
BLACK = (0, 0, 0)
WHITE = (255, 255, 255)
RED = (255, 0, 0)

# initialization
pygame.init()

# load image (background, enemy, buttons)     匯入所需圖片，並調整成要的大小
background_image = pygame.transform.scale(pygame.image.load("images/Map.png"), (WIN_WIDTH, WIN_HEIGHT))
enemy_image = pygame.transform.scale(pygame.image.load('images/enemy.png'),(60,60))
hp_image = pygame.transform.scale(pygame.image.load('images/hp.png'),(HP_WIDTH,HP_HEIGHT))
hp_gray_image = pygame.transform.scale(pygame.image.load('images/hp_gray.png'),(HP_WIDTH,HP_HEIGHT))
coutinue_image = pygame.transform.scale(pygame.image.load('images/continue.png'),(BTN_WIDTH,BTN_HEIGHT))
muse_image = pygame.transform.scale(pygame.image.load('images/muse.png'),(BTN_WIDTH,BTN_HEIGHT))
pause_image = pygame.transform.scale(pygame.image.load('images/pause.png'),(BTN_WIDTH,BTN_HEIGHT))
sound_image = pygame.transform.scale(pygame.image.load('images/sound.png'),(BTN_WIDTH,BTN_HEIGHT))

class Game:
    def __init__(self):
        # window    新增一個surface叫做win
        self.win = pygame.display.set_mode((WIN_WIDTH, WIN_HEIGHT))
        #fill your window in white
        self.win.fill(WHITE)        
        # set the title
        pygame.display.set_caption("My first game")
        #time    設定左下方要顯示時間的字體及大小(字體 None表示為預設字體)
        self.font = pygame.font.SysFont(None, 24)
        # hp
        self.hp = 7
        self.max_hp = 10
        pass

    def game_run(self):         # game loop
        start_time = pygame.time.get_ticks()      #紀錄遊戲開始時的內部時間
        run = True
        while run:   #只要沒有關閉視窗，就會在while迴圈中繼續遊戲畫面
            # event loop
            for event in pygame.event.get():
                if event.type == pygame.QUIT:    #按叉叉關閉視窗時，將run設為False，跳出迴圈，關閉視窗
                    run = False
                
            #draw background
            self.win.blit(background_image, (0, 0))    #顯示背景圖(地圖)
            
            # draw enemy and health bar
            self.win.blit(enemy_image, (enemy_x, enemy_y))    #顯示enemy           
            pygame.draw.rect(self.win, RED, [enemy_x+3, enemy_y, 50, 5])  #血條顯示在enemy位置的上方一些

            # draw menu (and buttons)
            pygame.draw.rect(self.win, BLACK, [0,0,WIN_WIDTH,80])     #上方顯示黑色橫幅
            self.win.blit(muse_image, (700,0))           #此四行為顯示右上角四個按鈕
            self.win.blit(sound_image, (780,0))
            self.win.blit(coutinue_image, (860,0))
            self.win.blit(pause_image, (940,0))
            
            hp_x1 = 400   #第一行hp值顯示的x座標
            hp_x2 = 400   #第二行hp值顯示的x座標
            hp_y1 = 0        #第一行hp值的y座標
            hp_y2 = 40       #第二行hp值的y座標
            
            for hp_count in range (self.max_hp):     #小於最大hp時進入迴圈，顯示出血量
                if hp_count < 5:              #第一行hp值
                    if hp_count < self.hp:    #判斷血量被扣了沒
                        self.win.blit(hp_image, (hp_x1, hp_y1))
                    else:                                         #被扣了
                        self.win.blit(hp_gray_image, (hp_x1, hp_y1))
                    hp_x1 += 40                                   #x座標右移
                    hp_count += 1
                else:                          #第二行hp值
                    if hp_count < self.hp:     #判斷血量被扣了沒
                        self.win.blit(hp_image, (hp_x2, hp_y2))
                    else:                                         #被扣了
                        self.win.blit(hp_gray_image, (hp_x2, hp_y2))
                    hp_x2 += 40                                   #座標右移 
                    hp_count += 1

            time = (pygame.time.get_ticks() - start_time)//1000   #取得現在的內部時間，與初始時間相減得目前的秒數(因為取得的是毫秒，所以除1000取整數)
            minute = str(time // 60)                   #轉換成分，並轉成str
            second = str(time % 60)                    #轉換成秒，並轉成str
            #if len(minute)<2:                          #分鐘只有個位數的時候補零給它顯示
                #minute = '0' + minute                  #本來是想要對稱顯示，但看助教給的範例，分鐘只有顯示一位數，所以註解掉了
            if len(second)<2:                          #秒只有個位數的時候補零給它顯示
                second = '0' + second
            # draw time
            text_surface = (self.font.render(minute + ':' + second, True, WHITE, BLACK))   #建立要顯示的時間文字surface
            pygame.draw.rect(self.win, BLACK, [0, 580, 50, 20])
            self.win.blit(text_surface, (3,583))       #顯示時間區塊在左下角
            pygame.display.update()
        pygame.quit()

if __name__ == "__main__":

    covid_game = Game()
    covid_game.game_run()
```
